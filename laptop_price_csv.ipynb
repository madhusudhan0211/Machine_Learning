```python
find_data = '512GB'
```

```python
df['512GB'] = df['Memory'].str.find(find_data)
```

```python
df.head()
```

```python
x = df.pop('512GB')
```

```python
x = x.replace(to_replace = 0 , value = 1)
```

```python
x = x.replace(to_replace = -1 , value = 0)
```

```python
x
```

```python
df['512GB'] = x
df.head()
```
